## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# droping the unnecessary columns from the dataset :
data.drop(['name','host_name','id','last_review'],axis=1,inplace=True)
```

```python
# removing the null values from review_per_month column:
data.fillna({'reviews_per_month':0},inplace=True)
data.head()
```

```python
# ploting distplot to check the distribution of price :
sns.displot(data.price)
plt.show()
```

```python
# checking for outliers in price column :
sns.boxplot(data.price)
plt.show()
```

* AS we can see there are many outliers and we need to remove this outliers 

```python
#using inter quantile range approach to remove outliers.
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data = data[(data['price'] >= lower_bound) & (data['price'] <= upper_bound)]
```

```python
sns.boxplot(data.price)
plt.show()
```

```python
# ploting distplot to check the distribution of price after removing outliers  :
sns.displot(data.price)
plt.show()
```

### What all manipulations have you done and insights you found?

First i removed the unnecessary colunms such as name, host_name, id, last_review.
Then filling the null values in the colunm of reviews_per_month with 0.
After that it was time for removing outliers for removing outliers first i confirmed the outliers by ploting boxplot and then used inter quntile range to remove those outliers. now our dataset is ready for Exploratory data analysis.

# Q-1 Which neighborhood group are the most popular for Airbnb listings?

```python
# checking for neighbourhood group which are most popular for airbnb listings :
most = data.groupby('neighbourhood_group')['host_id'].count()
most = most.sort_values(ascending = False)
most
```

```python
# ploting neighbourhood group with most listing:
plt.figure(figsize=(8,5))
plt.title("Neighbourhood Group")
g = plt.pie(data.neighbourhood_group.value_counts(), labels=data.neighbourhood_group.value_counts().index,autopct='%1.1f%%')
plt.show()
```

# Q-2 What are the most common types of properties listed on Airbnb?

```python
# most common properties listed on airbnb: 
common_type = data.groupby('room_type')['host_id'].count()
common_type = common_type.sort_values(ascending=False)
common_type
```

```python
# ploting the above result :
plt.bar(common_type.index, common_type.values)
plt.title('Top 5 Most Common Property Types on Airbnb')
plt.xlabel('Property Type')
plt.ylabel('Number of Listings')
plt.show()
```

* As we can see the Entire home/apt and private room  are high in numbers.

* From neighbourhood group we get to know that Manhattan and broklyn has most of the listings and staten island has lowest number of rooms.

# Q-3 Top neighbourhoods in NYC with respect to average price per day of Airbnb listings?

```python
df = data.groupby(['neighbourhood_group','neighbourhood'])['price'].mean().reset_index()
top_neigh = df.loc[df.groupby('neighbourhood_group')['price'].idxmax()].sort_values(by='price',ascending=False)
top_neigh
```

```python
top_neigh.plot.bar(x='neighbourhood', rot=90, title='Top Neighbourhoods in each neighbourhood groups in NYC')
plt.xlabel('neighbourhoods')
plt.ylabel('price')
plt.show()
```

* Among the top neighbourhoods in each neighbourhood groups, top 2 of them namely: Fort Wadsworth & Sea Gate, origins from Staten Island & Brooklyn.

Riverdale from Bronx and Tribeca from Manhattan. Neponsit from Queens.

# Q - 4 What factors have the biggest impact on the price of an Airbnb listing?

```python
x= data.corr()
sns.heatmap(x, annot=True)
plt.show()
```

* Observations on price:
 * price seems to be positively correlated with longitude meaning that one can expect higher prices as position in NYC moves West. This is expected because Manhattan, which is the most expensive borough of the city, is located on the west side of the city
 * latitude seems to have lesser effect on the price. However, there is a slight indication of higher prices located in the northern parts of the city
price is also positively correlated with: increasing availability, the fact that the property is rented by a host who lists other properties, and increasing number of minimum nights
 * price is negatively correlated with number of reviews and reviews per month, indicating that it is possible that the prior reviews could depress the prices to some extent
